```python
import easyocr
import cv2

image = cv2.imread("handwritten_text.png")

reader = easyocr.Reader(["en"])

results = reader.readtext(image)

for r in results:
    print(r[1])

words = []

for r in results:
    words.append(r[1])

sentence = " ".join(words)
print(sentence)
```

```python
import easyocr
import cv2
import re

image = cv2.imread("bill2.png")
reader = easyocr.Reader(["en"])

items = []
prices = []

reader.readtext(image)

results = reader.readtext(image)


for result in results: 
    text = result[1]

    if re.fullmatch(r"^[A-Za-z]+(_?\d+)?$", text):
        items.append(text)
    if re.fullmatch(r"\d+(\.\d+)?", text):
        prices.append(float(text))
print(sum(prices))
print(items)
```
